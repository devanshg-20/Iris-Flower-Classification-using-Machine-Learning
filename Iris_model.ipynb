```python
# Importing Dataset
import pandas as pd
df = pd.read_csv("Iris.csv")
df.info()
df.head()
```

```python
# Importing Essential Libraries
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Dividing the data into X and y
X = df.drop(["Species","Id"], axis=1)
y = df["Species"]
X
y
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Logistic Regression 
Logistic_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model1', LogisticRegression(max_iter=200))
])

param_grid = {"model1__C" : [0.01,0.1,1,10]}

LRclassifierCV = GridSearchCV(
    Logistic_pipeline,
    param_grid,
    cv = 5
)

LRclassifierCV.fit(X_train,y_train)
y_pred = LRclassifierCV.predict(X_test)

print("Best LR Params:", LRclassifierCV.best_params_)
print("Best LR Score:", LRclassifierCV.best_score_)
print("Logistic Regression Accuracy: ", accuracy_score(y_test,y_pred))
print("Confusion Matrix\n", confusion_matrix(y_test,y_pred))
print("Classification Report\n", classification_report(y_test,y_pred))
```

```python
# K Nearest Neighbors
K_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model2', KNeighborsClassifier())
])

param_grid = {"model2__n_neighbors" : [3,5,7,9]}

KNNclassifierCV = GridSearchCV(
    K_pipeline,
    param_grid,
    cv = 5
)

KNNclassifierCV.fit(X_train,y_train)
y_pred = KNNclassifierCV.predict(X_test)


print("Best KNN Params:", KNNclassifierCV.best_params_)
print("Best KNN Score:", KNNclassifierCV.best_score_)
print("KNN Accuracy: ", accuracy_score(y_test,y_pred))
print("Confusion Matrix\n", confusion_matrix(y_test,y_pred))
print("Classification Report\n", classification_report(y_test,y_pred))
```

```python
# Naive Bayes
Naive_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model3', GaussianNB())
])

param_grid = {"model3__var_smoothing" : [1e-9,1e-8,1e-7]}

NBclassifierCV = GridSearchCV(
    Naive_pipeline,
    param_grid,
    cv = 5
)

NBclassifierCV.fit(X_train,y_train)
y_pred = NBclassifierCV.predict(X_test)


print("Best NB Params:", NBclassifierCV.best_params_)
print("Best NB Score:", NBclassifierCV.best_score_)
print("Naive Bayes Accuracy", accuracy_score(y_test,y_pred))
print("Confusion Matrix\n", confusion_matrix(y_test,y_pred))
print("Classification Report\n", classification_report(y_test,y_pred))
```

```python
# Out of the Predictions
samples = [
    [5.1,3.5,1.4,0.2],
    [6.0,2.9,4.5,1.5],
    [6.5,3.0,5.5,2.0]
]

predictions = NBclassifierCV.best_estimator_.predict(samples)

print(predictions)
```
